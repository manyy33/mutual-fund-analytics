```python
import pandas as pd

# Step 1: Load
df_nav = pd.read_csv(r'C:\Users\user\Desktop\mutual_fund_analytics\data\raw\02_nav_history.csv')

# Step 2: Convert date
df_nav['date'] = pd.to_datetime(df_nav['date'], errors='coerce')
df_nav = df_nav.dropna(subset=['date'])

# Step 3: Validate NAV > 0
df_nav = df_nav[df_nav['nav'] > 0]

# Step 4: Sort
df_nav = df_nav.sort_values(by=['amfi_code', 'date']).reset_index(drop=True)

# Step 5: Forward fill per fund
df_nav['nav'] = df_nav.groupby('amfi_code')['nav'].ffill()

# Step 6: Remove duplicates
df_nav = df_nav.drop_duplicates()

# Confirm
print("Shape:", df_nav.shape)
print("Columns:", df_nav.columns.tolist())
print("Missing values:\n", df_nav.isnull().sum())
print("\nFirst 5 rows:\n", df_nav.head())
```

```python
df_nav.to_csv(r'C:\Users\user\Desktop\mutual_fund_analytics\data\processed\nav_history_cleaned.csv', index=False)

print("✅ File saved successfully!")
print("Final Shape:", df_nav.shape)
print("Date range:", df_nav['date'].min(), "to", df_nav['date'].max())
print("NAV min:", df_nav['nav'].min(), "| max:", df_nav['nav'].max())
print("Unique funds:", df_nav['amfi_code'].nunique())
```
